```python
from tensorflow.keras.layers import Dense, Flatten , Input
from tensorflow.keras.models import Model
from tensorflow.keras.preprocessing import image
from tensorflow.keras.preprocessing.image import ImageDataGenerator ,load_img
from tensorflow.keras.applications.vgg16 import VGG16 , preprocess_input
from glob import glob
import numpy as np
import matplotlib.pyplot as plt
```

```python
vgg = VGG16(input_shape=(224,224,3),weights = 'imagenet',include_top=False)
```

```python
for layer in vgg.layers:
  layer.trainable = False
```

```python
x = Flatten()(vgg.output)
```

```python
prediction = Dense(17,activation='softmax')(x)
```

```python
model = Model(inputs = vgg.input,outputs=prediction)
```

```python
model.summary()
```

```python
model.compile(loss = 'categorical_crossentropy',optimizer = 'adam',metrics=['accuracy'],run_eagerly=True)
```

```python
train_datagen = ImageDataGenerator(rescale = 1./255 , shear_range = 0.2 , zoom_range = 0.2 , horizontal_flip = True)

test_datagen = ImageDataGenerator(rescale = 1./255)
```

```python

train_path = "/content/Dataset/train"
test_path = "/content/Dataset/test"
```

```python

train = train_datagen.flow_from_directory(train_path,
                                      target_size=(224,224),
                                      batch_size=22,
                                      class_mode='categorical')

test = test_datagen.flow_from_directory(test_path,
                                      target_size=(224,224),
                                      batch_size=22,
                                      class_mode='categorical')
```

```python
import sys
r = model.fit(train, validation_data=test,epochs = 100 , steps_per_epoch = len(train)//3 , validation_steps=len(test)//3)
```

```python
model.save('mymodel.h5')
```
